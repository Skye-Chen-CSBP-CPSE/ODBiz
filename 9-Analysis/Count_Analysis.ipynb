```python

import pandas as pd
import numpy as np
import math


data = pd.read_csv('/home/jovyan/ODBiz/8-Cleaning/ODBiz_Final.csv', low_memory=False)



# how many total rows there are in the dataset
count_unique_idx = data['idx'].nunique()     
print(f'There are a total of {count_unique_idx} unique rows in this dataset')



#how many NAICS entries do we have in total 
count_unique_naics = data['derived_NAICS'].count()   
print(f'There are a total of {count_unique_naics} rows in derived_NAICS')


#how many NAICS entries were gviven to us 
count_unique_naics = data['source_NAICS_primary'].count()   
print(f'There are a total of {count_unique_naics} rows in source_NAICS_primary')


#how many NAICS codes we found 
diff = data['derived_NAICS'].count() - data['source_NAICS_primary'].count()
print(f'That means that we found a total of {diff} NAICS codes ourselves')


#how many NAICS codes are missing
diff_2 = data['idx'].nunique() -  data['derived_NAICS'].count() 
print(f'That means that we are missing a total of {diff_2} NAICS codes')


#how many total businesses by province
group_prov_idx = data.groupby('province')['idx'].count() 
print('Total businesses count by province')
print(group_prov_idx)




#how many provinces by naics groups (this is Total NAICS(derived_NAICS))
group_prov_naics = data.groupby('province')['derived_NAICS'].count() 
print('How many total NAICS in each province')
print(group_prov_naics)





#how many provinces by naics (this is Accurate NAICS(source_NAICS_primary))
group_naics =data.groupby('province')['source_NAICS_primary'].count()
print('how many given NAICS in each Province ')
print(group_naics)



# How many given NAICS codes by province
group_naics_prov =data.groupby('source_NAICS_primary')['province'].count()
print('given NAICS count by Province ')
print(group_naics_prov)




# How many Total NAICS codes by province
group_naics_group_prov =data.groupby('derived_NAICS')['province'].count()
print('Total NAICS count per province ')
print(group_naics_group_prov)
```

```python

#indigenous count 

# how many total indigenous businesses there are 
count_unique_indi = (data['indigenous'] == 'yes')
count_unique_indi_True = (data['indigenous'] == 'TRUE')
count = sum(count_unique_indi) + sum(count_unique_indi_True)
print(f'There are a total of {count} indigenous businesses in this dataset')



# how many indigenous businesses per province
group_indi_prov = data.groupby(['indigenous','province'])['province'].count()
print('Indigenous businesses in each province')
print(group_indi_prov)



# how many indigenous businesses per Naics codes
group_indi_NAICS = data.groupby(['indigenous','NAICS_total'])['NAICS_total'].count()
print('Indigenous businesses per Naics codes')
print(group_indi_NAICS)



```
