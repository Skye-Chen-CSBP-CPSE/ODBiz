```python

import pandas as pd
import numpy as np
import math


data = pd.read_csv('/home/jovyan/ODBiz/9-Analysis/1PLA.csv', low_memory=False)





# how many total rows there are
count_unique_idx = data['idx'].nunique()     
print(f'There are a total of {count_unique_idx} unique rows in this dataset')



#how many total unique rows in naics_group
count_unique_naics = data['NAICS_total'].nunique()   
print(f'There are a total of {count_unique_naics} unique rows in NAICS_Total')




#how many total businesses by province
group_prov_idx = data.groupby('province')['idx'].count() 
print('Total businesses count by province')
print(group_prov_idx)




#how many provinces by naics groups (this is Total NAICS)
group_prov_naics = data.groupby('province')['NAICS_total'].count() 
print('How many NAICS_total in each province')
print(group_prov_naics)





#how many provinces by naics (this is Accurate NAICS)
group_naics =data.groupby('province')['NAICS'].count()
print('how many Accurate NAICS in each Province ')
print(group_naics)



# How many Accurate NAICS codes by province
group_naics_prov =data.groupby('NAICS')['province'].count()
print('Accurate NAICS count by Province ')
print(group_naics_prov)




# How many Total NAICS codes by province
group_naics_group_prov =data.groupby('NAICS_total')['province'].count()
print('Total NAICS count per province ')
print(group_naics_group_prov)
```

```python

#indigenous count 

# how many total indigenous businesses there are 
count_unique_indi = (data['indigenous'] == 'yes')
count_unique_indi_True = (data['indigenous'] == 'TRUE')
count = sum(count_unique_indi) + sum(count_unique_indi_True)
print(f'There are a total of {count} indigenous businesses in this dataset')



# how many indigenous businesses per province
group_indi_prov = data.groupby(['indigenous','province'])['province'].count()
print('Indigenous businesses in each province')
print(group_indi_prov)



# how many indigenous businesses per Naics codes
group_indi_NAICS = data.groupby(['indigenous','NAICS_total'])['NAICS_total'].count()
print('Indigenous businesses per Naics codes')
print(group_indi_NAICS)



```
