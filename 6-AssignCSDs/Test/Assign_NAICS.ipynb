```python


import pandas as pd
import numpy as np
import math


df = pd.read_csv("/home/jovyan/ODBiz/6-AssignCSDs/Test/Test_CSDs3.csv")

# STEP 1
# copy over NAICS values that already exist in the data


# This fills all the null values in the columns with 10 for analysis
df.primary_NAICS.fillna(value=10, inplace=True)


def first_two(d):
    return (d // 10 ** (int(math.log(d, 10)) - 1))


# print new column with only first two digits
df['NAICS_Group'] = df.primary_NAICS.apply(first_two)


# returns the 10 values back to null
df['primary_NAICS'] = df['primary_NAICS'].replace(10, '')
df['NAICS_Group'] = df['NAICS_Group'].replace(10, np.nan)


# STEP 2
# find the NAICS numbers that
# were put in business_sector column


agri = '11'
mining = '21'
uti = '22'
cons = '23'
mani = '31-33'
whole = '41'
retail = '44-45'
trans = '48-49'
info = '51'
fin = '52'
real = '53'
prof = '54'
mang = '55'
admin = '56'
edu = '61'
health = '62'
arts = '71'
acco = '72'
other = '81'
pub = '91'


df.loc[df['business_sector'].str.contains(agri, case= False, na= False), 'NAICS_Group']= 11.0
df.loc[df['business_sector'].str.contains(mining, case = False, na = False), 'NAICS_Group' ] = 21.0
df.loc[df['business_sector'].str.contains(uti, case = False, na = False), 'NAICS_Group' ] = 22.0
df.loc[df['business_sector'].str.contains(cons, case = False, na = False), 'NAICS_Group' ] = 23.0
df.loc[df['business_sector'].str.contains(mani, case = False, na = False), 'NAICS_Group' ] = 31.0
df.loc[df['business_sector'].str.contains(whole, case = False, na = False), 'NAICS_Group' ] = 41.0
df.loc[df['business_sector'].str.contains(retail, case = False, na = False), 'NAICS_Group' ] = 44.0
df.loc[df['business_sector'].str.contains(trans, case = False, na = False), 'NAICS_Group' ] = 48.0
df.loc[df['business_sector'].str.contains(info, case = False, na = False), 'NAICS_Group' ] = 51.0
df.loc[df['business_sector'].str.contains(fin, case = False, na = False), 'NAICS_Group' ] = 52.0
df.loc[df['business_sector'].str.contains(real, case = False, na = False), 'NAICS_Group' ] = 53.0
df.loc[df['business_sector'].str.contains(prof, case = False, na = False), 'NAICS_Group' ] = 54.0
df.loc[df['business_sector'].str.contains(mang, case = False, na = False), 'NAICS_Group' ] = 55.0
df.loc[df['business_sector'].str.contains(admin, case = False, na = False), 'NAICS_Group' ] = 56.0
df.loc[df['business_sector'].str.contains(edu, case = False, na = False), 'NAICS_Group' ] = 61.0
df.loc[df['business_sector'].str.contains(health, case = False, na = False), 'NAICS_Group' ] = 62.0
df.loc[df['business_sector'].str.contains(arts, case = False, na = False), 'NAICS_Group' ] = 71.0
df.loc[df['business_sector'].str.contains(acco, case = False, na = False), 'NAICS_Group' ] = 72.0
df.loc[df['business_sector'].str.contains(other, case = False, na = False), 'NAICS_Group' ] = 81.0
df.loc[df['business_sector'].str.contains(pub, case = False, na = False), 'NAICS_Group' ] = 91.0


#df.to_csv('/home/jovyan/ODBiz/6-AssignCSDs/Test/Filtred_NAICS.csv',index=False)
df.to_csv('/home/jovyan/ODBiz/6-AssignCSDs/Test/3NAICS_Test.csv', index=False)
```
