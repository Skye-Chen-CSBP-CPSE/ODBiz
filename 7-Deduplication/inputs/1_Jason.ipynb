```python
'''
1. json-generator 

Creates JSON source files for opentabulate according to a variable map csv.
Modified for use with ODBiz data sources by Skye Chen
''' 

import csv
import json
from tqdm import tqdm

# Define file path name
# DeDupDir = '/home/jovyan/ODBiz/Deduplication/ODHF_copy/inputs'

# input variable map file
input_file = csv.DictReader(open('/home/jovyan/ODBiz/7-Deduplication/inputs/4NAICS_Test.csv', encoding = 'utf-8'))

# define schema objects (Forced for all features in data source)
metadataFields = ('localfile',)

geoFields = ('longitude', 'latitude')

bizFields = ('business_name',
        'business_sector',
        'business_subsector',
        'business_description',
        'business_id_no',
        'licence_number',
        'licence_type',
        'primary_NAICS',
        'secondary_NAICS',
        'NAICS_descr',
        'alt_econ_act_code',
        'alt_econ_act_descrip',
        'business_website',
        'email',
        'telephone',
        'telephone_extension',
        'toll_free_telephone',
        'fax',
        'total_no_employees',
        'no_full_time',
        'no_part_time',
        'no_seasonal',
        'date_established',
        'indigenous',
        'status')

addressFields = ('full_address',
                'full_address_2',
                'mailing_address',
                'unit',
                'street_no',
                'street_name',
                'street_direction',
                'street_type',
                'city',
                'province',
                'postal_code',
                'country')

for row in (input_file): # Each row is a data source

    #filename = f"{row['localfile']}"
    #json_filename = filename.replace(".csv", ".json")
    #OP = open(f"/home/jovyan/ODBiz/Deduplication/ODHF_copy/inputs/1NAICS_Test.csv/{json_filename}", "w")
    jsonFilePath = r'4NAICS_Test.json'
    OP = open(jsonFilePath, 'w', encoding='utf-8')
    #as jsonf:
       # jsonf.write(json.dumps)
     
    
    
    
    # Dictionaries for schema json field
    metadata_dict = {}
    geo_dict = {}
    address_dict = {}
    biz_dict = {}

    # Only adding k,v pairs that are not blank
    for f in metadataFields:
        if row[f] != '':
            metadata_dict[f] = row[f]
    for f in geoFields:
        if row[f] != '':
            geo_dict[f] = row[f]
    for f in addressFields:
        if row[f] != '':
            address_dict[f] = row[f]
    for f in bizFields:
        if row[f] != '':
            biz_dict[f] = row[f]
            
    # function to add force to a string

    # SL define json structure and static fields
    jsondict = {
        "filename":'inputs/1NAICS_Test.csv',
        "output_name":'outputs/created_pairs.csv',
        "encoding":"utf8",
        "index":"idx",
        "column_map":{
            "FileName":"localfile",
            "Name":"business_name",
            "NAICS":"NAICS_Group",
            "Date":"date_established",
            "Status":"status",
            "Address":"full_address",
            "StreetNumber":"street_no",
            "StreetName":"street_name",
            "PostalCode":"postal_code",	
            "City":"city",
            "Province":"province",
            "Latitude":"latitude",
            "Longitude":"longitude",
            "CSDUID":"CSDUID"
            },
        }
        
    # Create the dictionary for the json output and write to file
    json_data = json.dumps(jsondict, indent=4)  # Formatting
    OP.write(json_data)  # Write to json
    #print(f'File saved to {OP.name}')
    OP.close()

    
    
    
    
     
```
